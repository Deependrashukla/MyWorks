```python
adj_lst = {}

for key in nodes:
    lst = []
    for node1, node2, weight in zip(df['node1'], df['node2'], df['weight']):
        if key == node1:
            lst.append((node2, weight))

    adj_lst[key] = lst
```

```python
heulistic_dis = {}

for node, weight in zip(df2['node1'], df2['distance_to_Bucharest']):
    heulistic_dis[node] = weight
```

```python
path_lst = []

def GBFS(start_node, target_node):
    global path_lst

    if start_node not in path_lst:
        path_lst.append(start_node)
        if target_node in path_lst: 
            return

        num = float('inf')
        next_node = ''
        for i in adj_lst[start_node]:
            node = i[0]
            if heulistic_dis[node] < num:
                num = heulistic_dis[node]
                next_node = node

        GBFS(next_node, target_node)   

GBFS('Arad', 'Bucharest')
```
