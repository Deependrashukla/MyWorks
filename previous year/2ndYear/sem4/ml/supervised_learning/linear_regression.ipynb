```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("SelfShiksha_SLB_MCQ_80.xlsx")
df
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x1 = [int(item) for item in list(df['x'])]
x2 = [num**2 for num in list(df['x'])]

df["x1"] = x1
df["x2"] = x2

df.drop(['x'], axis=1, inplace = True)
```

```python
x = df.drop(['y'], axis=1)
y = df.y

reg = LinearRegression()
reg.fit(x,y)
```

```python
print(reg.coef_, reg.intercept_)
```
